```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
df=pd.read_csv("general_data.csv") 
df.head(2)
```

```python
df["NumCompaniesWorked"].fillna(df['NumCompaniesWorked'].mean(),inplace=True)
df["TotalWorkingYears"].fillna(df['TotalWorkingYears'].mean(),inplace=True)
df.drop(['EmployeeID', 'EmployeeCount','Over18','StandardHours'],axis=1,inplace=True)
```

```python
df['Attrition'].replace({'No': 0, 'Yes': 1},inplace = True)
df['Gender'].replace({'Female': 0, 'Male': 1},inplace = True)
df['Department'].replace({'Sales': 0, 'Research & Development': 1, 'Human Resources' : 2},inplace = True)
df['EducationField'].replace({'Life Sciences': 0, 'Other': 1, 'Marketing' : 2, 'Medical':3, 'Technical Degree':4, 'Human Resources' : 5 },inplace = True)
df['BusinessTravel'].replace({'Travel_Rarely': 3, 'Travel_Frequently': 2,'Non-Travel': 1},inplace = True)
```

```python
df.head(4)
```

```python
from scipy.stats import pearsonr
```

```python
stats,p=pearsonr(df.Attrition,df.BusinessTravel)
print(stats,p)
```

```python
stats,p=pearsonr(df.Attrition,df.Department)
print(stats,p)
```

```python
stats,p=pearsonr(df.Attrition,df.DistanceFromHome)
print(stats,p)
```

```python
st,q=pearsonr(df.Attrition,df.EducationField)
print(st,q)
```

```python
st3,g=pearsonr(df.Attrition,df.Education)
st4,m=pearsonr(df.Attrition,df.EducationField)
st5,n=pearsonr(df.Attrition,df.Gender)
st6,s=pearsonr(df.Attrition,df.NumCompaniesWorked)
st7,l=pearsonr(df.Attrition,df.PercentSalaryHike)
st8,gg=pearsonr(df.Attrition,df.TotalWorkingYears)
st9,mm=pearsonr(df.Attrition,df.TrainingTimesLastYear)
st10,nn=pearsonr(df.Attrition,df.YearsAtCompany)
st11,ss=pearsonr(df.Attrition,df.YearsSinceLastPromotion)
st12,ll=pearsonr(df.Attrition,df.YearsWithCurrManager)
```

```python
print(st3,g)
print(st4,m)
print(st5,n)
print(st6,s)
print(st7,l)
print(st3,gg)
print(st8,mm)
print(st9,nn)
print(st10,ss)
print(st11,ll)
```
